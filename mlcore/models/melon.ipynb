```python
from IPython.display import Image
from ultralytics import YOLO
```

```python
model = YOLO("yolo11m-cls")

result = model.train(
    data='/kaggle/input/melon-dataset/melon',
    epochs=500,
    imgsz=320, 
    optimizer='AdamW',

    degrees=10,
    shear=5,
    batch=32,

    save=True,
    plots=True,
    name='melons_cls_model',
)
```
