```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo11m-cls")

result = model.train(
    data='/kaggle/input/tomatodataset/tomato',
    epochs=400,
    imgsz=256, 
    optimizer="AdamW",
    save=True,
    plots=True,
    name='tomatoe_cls_model',
)
```
